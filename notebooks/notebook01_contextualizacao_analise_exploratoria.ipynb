# LogiMarket Logistics Analytics — Eficiência de Entregas e Satisfação do Cliente

## Contextualização

### Cenário de Negócio

A LogiMarket é uma plataforma de marketplace que conecta milhares de vendedores a clientes em todo o Brasil. Com um modelo de negócio baseado em intermediacao, a experiencia de entrega representa um dos principais pontos de contato entre a plataforma e o consumidor final.
Em um mercado cada vez mais competitivo, onde o prazo de entrega e a confiabilidade logistica sao fatores decisivos na escolha de onde comprar, a LogiMarket enfrenta desafios crescentes relacionados a operacao de fulfillment. A equipe de operacoes identificou sinais de problemas recorrentes na cadeia de entregas, incluindo:

    Atrasos significativos entre a data estimada e a data real de entrega
    Disparidades regionais notaveis no tempo de transporte
    Oscilacoes na satisfacao do cliente correlacionadas a experiencias logisticas negativas
    Impacto direto na reputacao da marca e na taxa de recompra

### Problema de Negócio

    Como reduzir atrasos nas entregas e melhorar a experiencia do cliente sem comprometer a eficiencia operacional?

A equipe de lideranca precisa de respostas baseadas em dados para responder as seguintes questoes estrategicas:

    Quais fatores estao impactando o tempo de entrega e a satisfacao dos clientes?
    Existem regioes ou categorias de produto com performance logistica critica?
    Qual a correlacao entre atrasos e avaliacoes negativas?
    Como otimizar a operacao logistica de forma inteligente e sustentavel?

### Objetivo do Projeto

Desenvolver uma analise exploratoria completa dos dados de pedidos, entregas e satisfacao do cliente da LogiMarket, culminando em um dashboard interativo no Power BI que permita:

    Monitorar KPIs logisticos em tempo real (taxa de atraso, tempo medio de entrega, NPS)
    Identificar gargalos operacionais por regiao, categoria e periodo
    Segmentar perfis de risco de atraso para acao proativa
    Apoiar a tomada de decisao da equipe de operacoes com dados confiaveis

### Tecnologias Utilizadas

    Python (pandas, numpy, matplotlib, seaborn, plotly) para analise exploratoria e tratamento de dados
    Power BI para construcao do dashboard interativo
    GitHub Pages para publicacao do portfolio
